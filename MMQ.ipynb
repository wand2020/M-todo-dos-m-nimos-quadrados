# **Método dos mínimos quadrados**

```python
# Importar bibliotecas para uso de algoritmos matemáticos
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definindo lista de pontos onde cada elemento da lista
# é um vetor de coordenadas x e y
pontos = np.array([
  # [x, y]
  [1, 1],
  [3, 2],
  [4, 4],
  [6, 4],
  [8, 5],
  [9, 7],
  [11, 8],
  [14, 9]
])
```

## Gerando gráfico para observar se há um comportamento linear

```python
# Definindo a função gráfico
# O único argumento extritamente necessário é a lista
# lista de pontos.
def grafico(
  pontos, 
  titulo_grafico="Gráfico Linear", 
  titulo_x="Medidas X", 
  titulo_y="Medidas Y",
  cor="black",
  linhas=False
):
  x_i = pontos[:, 0]
  y_i = pontos[:, 1]

  fig = plt.figure()
  ax = fig.add_axes([0, 0, 1, 1])
  ax.scatter(x = x_i, y = y_i, color = cor)
  plt.title(titulo_grafico)
  plt.xlabel(titulo_x)
  plt.ylabel(titulo_y)
  
  if linhas:
    plt.plot(x_i, y_i, color=cor)

  plt.show()
```

```python
# Chamando a função gráfico.
# Os argumentos devem ser colocados na ordem que 
# foram definidos acima!
grafico(pontos)
```

```python
# Gráfico conectando linhas e pontos
grafico(pontos, linhas=True)
```

## Entendendo um comportamento próximo do linear aplica-se o MMQ

```python
# Número de pontos
n = len(pontos)

x_i = pontos[:, 0]
y_i = pontos[:, 1]

x_i_pow2 = x_i ** 2
x_i_times_y_i = x_i * y_i

# Somatorios
somatorio_x = sum(x_i)
somatorio_y = sum(y_i)
somatorio_xy = sum(x_i_times_y_i)
somatorio_xpow2 = sum(x_i_pow2)

# Cálculando os coeficientes a e b pelo MMQ
a = (somatorio_x * somatorio_y - n * somatorio_xy) / (somatorio_x ** 2 - n * somatorio_xpow2)
b = (somatorio_xy * somatorio_x - somatorio_xpow2 * somatorio_y) / (somatorio_x ** 2 - n * somatorio_xpow2)

print(
  f"""
  Somatorio x = {somatorio_x}
  Somatorio y = {somatorio_y}
  Somatorio xy = {somatorio_xy}
  Somatorio x^2 = {somatorio_xpow2}

  a = {"{:.2f}".format(a)}
  b = {"{:.2f}".format(b)}
  """
)
```

```python
# Criando lista de pontos ajustados
y_ajustado = [a * x + b for x in pontos[:, 0]]
print(y_ajustado)
```

## GRÁFICO AJUSTADO

```python
def graficoAjustado(
  pontos,
  y_ajustado, 
  titulo_grafico="Gráfico Ajustado", 
  titulo_x="Medidas X", 
  titulo_y="Medidas Y",
  cor="black",
  cor_secundaria="red",
  linhas=True
):
  x_i = pontos[:, 0]
  y_i = pontos[:, 1]

  fig = plt.figure()
  ax = fig.add_axes([0, 0, 1, 1])
  ax.scatter(x = x_i, y = y_ajustado, color = cor)
  ax.scatter(x = x_i, y = y_i, color = cor_secundaria)
  plt.title(titulo_grafico)
  plt.xlabel(titulo_x)
  plt.ylabel(titulo_y)
  
  if linhas:
    plt.plot(x_i, y_ajustado, color=cor)

  plt.show()
```

```python
# Gerando gráfico com as medidas e comparando com o ajuste
graficoAjustado(pontos, y_ajustado)
```
